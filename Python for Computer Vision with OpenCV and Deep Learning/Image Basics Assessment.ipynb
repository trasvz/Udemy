```python
#Task 1 : open image in folder and correct the RGB order
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('img/IMG20170525134042.jpg')
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#Task 2 : Flip the image upside down
```

```python
img_flip = cv2.flip(img, 0)
plt.imshow(img_flip)
```

```python
#Task 3 : Draw and empty red rectangle in the face
```

```python
img_draw = img.copy()
```

```python
cv2.rectangle(img_draw,pt1=(250,100),pt2=(900,700),color=(255,0,0),thickness=5)
plt.imshow(img_draw)
```

```python
#Task 4 : Draw blue triangle in the face
```

```python
img_draw2 = img.copy()
vertices = np.array([[650,25], [900,700], [250,700]],dtype=np.int32)
vertices = vertices.reshape((-1,1,2))
cv2.polylines(img_draw2,[vertices],isClosed=True,color=(0,0,255),thickness=5)
plt.imshow(img_draw2)
```

```python
#Bonus Task : Fill the triangle
```

```python
cv2.fillPoly(img_draw2,[vertices], color=(0,0,255))
plt.imshow(img_draw2)
```

```python
#Bonus Task : Draw rectangle just by right clicking
```

```python
import cv2
import numpy as np

def draw_cir(event,x,y,flags,param):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,0,255),0)

cv2.namedWindow(winname='mydraw')
cv2.setMouseCallback('mydraw',draw_cir)

# img = np.zeros((1024,1024,3))

while True:
    cv2.imshow('mydraw',img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
