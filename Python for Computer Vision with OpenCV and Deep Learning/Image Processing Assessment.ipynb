```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

### Task 1 : Open and display image that is located in the DATA folder

```python
img = cv2.imread('DATA/IMG20170525134042.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
display_img(img)
```

### Task 2 : Apply a binary threshold onto the image

```python
img = cv2.imread('DATA/IMG20170525134042.jpg',0)
img.min()
```

```python
ret, threshold_img = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
display_img(threshold_img,cmap='gray')
```

### Task 3 : Open the image from DATA folder and convert its colorspace to hsv

```python
img = cv2.imread('DATA/IMG20170525134042.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
display_img(img)
```

### Task 4 : Create a low pass filter with a 4 by 4 kernel filled with values of 1/10 (0.1) and then use 2D convolution to blur the image

```python
img = cv2.imread('DATA/IMG20170525134042.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# display_img(img)
```

```python
kernel = np.ones(shape=(4,4),dtype=np.float32)/10
kernel
```

```python
result = cv2.filter2D(img,-1,kernel)
display_img(result)
```

### Task 5 : Create a horizontal sobel filter (sobelx from lecture) with a kernel size of 5 to the grayscale version of the img and then display the resulting gradient filtered version of the img

```python
img = cv2.imread('DATA/IMG20170525134042.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# display_img(img)
```

```python
result = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
display_img(result)
```

### Task 6 : Plot the color histograms for the RED, BLUE, and GREEN channel of the img. Pay careful attention to the ordering of the channels

```python
img = cv2.imread('DATA/IMG20170525134042.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
display_img(img)
```

```python
color = ('r','g','b')

for i,col in enumerate(color):
    histo = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histo, color=col)
    plt.xlim([0,256])

plt.title('Histogram for the img')
```

```python
img = cv2.imread('DATA/IMG20170525134042.jpg')
color = ('b','g','r')

for i,col in enumerate(color):
    histo = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histo, color=col)
    plt.xlim([0,256])

plt.title('Histogram for the img')
```
